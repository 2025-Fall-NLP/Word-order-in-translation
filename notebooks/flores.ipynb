```python
# requirements

# pip install datasets
from datasets import load_dataset
```

```python
# FLORES-200 전체 데이터 불러오기
dataset = load_dataset("facebook/flores", "all")

# dataset은 dev(997), devtest(1012)로 나뉘어져 있음
# dev set에서 한 문장 샘플 선택 (인덱스 0 사용)
sample_sentence = dataset["dev"][0]
```

```python
# test) 4개 언어 선택
# 앞에 sentence 붙여야 됨
langs = {
    "English": "sentence_eng_Latn",
    "Korean": "sentence_kor_Hang",
    "Japanese": "sentence_jpn_Jpan",
    "Chinese (Simplified)": "sentence_zho_Hans",
}
```

```python
# 테스트 결과 출력
print("<Sample sentence from FLORES-200 (multilingual parallel sentence)>\n")

for name, code in langs.items():
    print(f"[{name}]")
    print(sample_sentence[code])
    print()
```
